```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit  # import the curve fitting function
%matplotlib inline
```

# Travel Time in Air

```python
plt.figure(figsize=(5,3));
plt.scatter(Time,Distance);

plt.ylabel('Path Length (m)',fontsize=20);
plt.xlabel('Time (s)',fontsize = 20);
plt.xticks(size = 13);
plt.yticks(size = 13);
plt.xlim(17e-9,41.5e-9);
#plt.ylim(0,1.2e-28);
```

```python
def myfun(t,c,k):
    ans =  c*t + k #Distance traveled with offset
    return ans

p0 = [3e8,-1e2] #guess

xspace = np.linspace(17e-9,41.5e-9) 

plsq, pcov = curve_fit(myfun, Time, Distance, p0)  # curve fit returns p and covariance matrix
# these give the parameters and the uncertainties
c = plsq[0]
ec = np.sqrt(pcov[0,0])
k = plsq[1]
ek = np.sqrt(pcov[1,1])


yfit = myfun(xspace,plsq[0],plsq[1])  # use fit results for a, b, c

plt.figure(figsize=(10,6));

plt.scatter(Time,Distance, marker='.',label='Data');
plt.plot(xspace,yfit,label='Fit')

plt.xlim(17e-9,41.5e-9);

plt.legend(loc='lower right')
plt.ylabel('Path Length (m)',fontsize=20);
plt.xlabel('Time (s)',fontsize = 20);
plt.text(1.8e-8,7,'$c = (%.2f \pm %.2f)\cdot 10^8$ m/s' % (plsq[0]*1e-8, np.sqrt(pcov[0,0])*1e-8),size=20)

#plt.errorbar(mub*B_Rb87,E_Rf,xerr = mub*16*muo*N*dV/(np.sqrt(125)*2*R),fmt='none');
plt.yticks(size = 15);
plt.xticks(size = 15);
#plt.savefig('Part1.png')
```
