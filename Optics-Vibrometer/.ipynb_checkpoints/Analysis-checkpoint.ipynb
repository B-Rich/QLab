```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit  # import the curve fitting function
%matplotlib inline
```

```python
f = np.array([50,63,74,85,98,108,124,151,77,83,70,96]) #Driving frequency


#Number of photodetector light/dark cycles in time (1/2f)
WavNum = np.array([20,25,42,42,33,20,12,6,42,49,33,38])


#Resonance (83Hz) Central peak separation time
ResSep = 82*1e-6 #[s]


wavelength = 633*1e-9 #633nm laser
```

# Excursion Distance

```python
ExcDist = wavelength*WavNum #Speaker excursion distance [m] 
ExcDist*1e6
```

```python
len(ExcDist)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(f,ExcDist*1e6)
plt.xlabel('Driving Frequency [Hz]',size=20);
plt.ylabel('Excursion Distance  [$\mu$m]', size = 20);
plt.xticks(size=12);
plt.yticks(size=12);
plt.savefig('zvf.png')
```

As the speaker travels, it changes the path length for light traveling in the speaker arm, thus the number of light/dark cycles between the speaker turning points gives the speaker excursion distance as an integral number of the laser beam wavelength.

To find the speaker average speed, we use the excursion distance and the value 1/2f, which represents the time spent between turning points. 

# Average Speed

```python
f
```

```python
t = np.array([0.5*1/entry for entry in f])

v = ExcDist/t #[m/s]
v*1e3 #mm/s
```

```python
plt.figure(figsize=(10,6))
plt.scatter(f,v*1e3)
plt.xlabel('Driving Frequency [Hz]',size=20);
plt.ylabel('Average Speed  [mm/s]', size = 20);
plt.xticks(size=12);
plt.yticks(size=12);
plt.savefig('vvf.png')
```
